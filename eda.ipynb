```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("clean_weather_data.csv")
df.head()
```

```python
df["temp_celsius"].hist()
plt.title("Temperature Distribution")
plt.xlabel("Temperature Celsius")
plt.ylabel("Count")
plt.show()
```

```python
df.plot(kind="bar", x="city", y="aqi")
plt.title("AQI by City")
plt.xlabel("City")
plt.ylabel("AQI")
plt.show()
```

```python
df.plot(kind="scatter", x="humidity_pct", y="aqi")
plt.title("Humidity vs AQI")
plt.xlabel("Humidity %")
plt.ylabel("AQI")
plt.show()
```
